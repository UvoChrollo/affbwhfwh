```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

#### 1 - Highlighting your data

```python
url = "https://assets.datacamp.com/production/repositories/3841/datasets/a6b11493e11dd47f3e03e0b96e2a2dbc51f03cb2/pollution_wide.csv"

pollution = pd.read_csv(url)
```

```python
pollution.head()
```

```python
houston = pollution.query("city == 'Houston'")
```

```python
houston.head()
```

```python
max_co2 = houston[houston['CO'] == houston['CO'].max()]
max_co2
```

```python
# Memberikan warna merah pada data dimana CO berada pada titik maksimal

houston_colors = ['red' if (day == 350) & (year == 2013) 
                  else 'black' for day, year in zip(houston.day, houston.year)]

# Set Figure Size
plt.rcParams['figure.figsize'] = (15,8)

# Main plot
ax = sns.scatterplot(x='NO2', y='SO2', data=houston,c=houston_colors)

# Horizontal Line
ax.axhline(8.425, linestyle='--', linewidth=1, color='red')

# Vertical Line
ax.axvline(54.1, linestyle='--', linewidth=1, color='red')

# Title
ax.set_title("Pengaruh NO2 terhadap SO2")

plt.show()
```
