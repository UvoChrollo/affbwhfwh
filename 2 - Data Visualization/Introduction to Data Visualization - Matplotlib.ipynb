```python
import matplotlib.pyplot as plt
import pandas as pd
```

#### 1 - Introduction to Matplotlib

```python
x_axis = [*range(1,11)]
y_axis = [i*2 for i in x_axis]
```

<b>Scatterplot</b>

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
plt.show()
```

<b>Line Plot</b>

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show()
```

<b>Bar Plot</b>

```python
data_a = [1,2,3,4]
data_b = [2,4,6,8]

fig, ax = plt.subplots()
ax.bar(data_a, data_b)
plt.show()
```

<b>DataFrame Visualization</b>

```python
link = "https://raw.githubusercontent.com/SophrosyneEunoia/Shockwave/main/seattle_weather.csv"

seattle = pd.read_csv(link)
seattle.head()
```

```python
seattle = seattle[seattle['STATION']=='USW00094290']
```

```python
seattle.head()
```

```python
seattle.shape
```

```python
x_axis = seattle['DATE']
y_axis = seattle['MLY-PRCP-NORMAL']

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='r', linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='r', linestyle='--', marker='d')
plt.show()
```

```python
data_45 = seattle['MLY-CLDD-BASE45']
data_50 = seattle['MLY-CLDD-BASE50']
data_55 = seattle['MLY-CLDD-BASE55']
data_57 = seattle['MLY-CLDD-BASE57']
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(x_axis, data_45)
ax[1].plot(x_axis, data_50)
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x_axis, data_45)
ax[0,1].plot(x_axis, data_50)
ax[1,0].plot(x_axis, data_55)
ax[1,1].plot(x_axis, data_57)
plt.show()
```

```python
fig, ax = plt.subplots(2,1,sharex=True)
ax[0].plot(x_axis, data_45)
ax[1].plot(x_axis, data_50)
plt.show()
```

#### 2 - Plotting time-series

```python
cli_link = "https://raw.githubusercontent.com/SophrosyneEunoia/Shockwave/main/climate_change.csv"

climate = pd.read_csv(cli_link)
climate.head()
```

```python
climate['date'] = pd.to_datetime(climate['date'])
climate.set_index('date',inplace=True)
climate.head()
```

```python
fig, ax = plt.subplots()

ax.plot(climate.index, climate['relative_temp'])

plt.show()
```

```python
fig, ax = plt.subplots()

seventies = climate["1970-01-01":"1979-12-31"]

ax.plot(seventies.index, seventies["co2"])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(climate.index, climate['relative_temp'])

ax2 = ax.twinx()

ax2.plot(climate.index, climate['co2'])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(climate.index, climate['relative_temp'], color='red')

ax2 = ax.twinx()

ax2.plot(climate.index, climate['co2'], color='blue')

plt.show()
```

#### 3 - Quantitative comparisons and statistical visualizations

```python
medal = "https://raw.githubusercontent.com/SophrosyneEunoia/Shockwave/main/medals_by_country_2016.csv"

medals = pd.read_csv(medal)
medals.head()
```

```python
medals.rename(columns={'Unnamed: 0':'Country'},inplace=True)
medals.set_index('Country',inplace=True)
medals.head()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
plt.xticks(medals.index, rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Silver'])
plt.xticks(medals.index, rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Bronze'])
plt.xticks(medals.index, rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'],bottom=medals['Gold'])
plt.xticks(medals.index, rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'],bottom=medals['Gold'])
ax.bar(medals.index, medals['Bronze'],bottom=medals['Gold']+medals['Silver'])
plt.xticks(medals.index, rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'],label="Gold")
ax.bar(medals.index, medals['Silver'],bottom=medals['Gold'],label="Silver")
ax.bar(medals.index, medals['Bronze'],bottom=medals['Gold']+medals['Silver'],label="Bronze")
plt.xticks(medals.index, rotation=90)
ax.legend()
plt.show()
```

```python
url = "https://raw.githubusercontent.com/SophrosyneEunoia/Shockwave/main/summer2016.csv"
summer = pd.read_csv(url)

men = summer['Sex'] == 'M'
sport = summer['Sport'] == 'Rowing'
mens_rowing = summer[men & sport]
```

```python
fig, ax = plt.subplots()
# Plot a histogram of "Weight" for mens_rowing
ax.hist(mens_rowing["Weight"])

# Set the x-axis label to "Weight (kg)"
ax.set_xlabel("Weight (kg)")

# Set the y-axis label to "# of observations"
ax.set_ylabel("# of observations")

plt.show()
```

```python
fig, ax = plt.subplots()

# Plot a histogram of "Weight" for mens_rowing
ax.hist(mens_rowing["Weight"], histtype='step', label="Rowing", bins=5)

ax.set_xlabel("Weight (kg)")
ax.set_ylabel("# of observations")

# Add the legend and show the Figure
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

# Add a bar for the rowing "Height" column mean/std
ax.bar("Rowing", mens_rowing["Height"].mean(), yerr=mens_rowing["Height"].std())

# Label the y-axis
ax.set_ylabel("Height (cm)")

plt.show()
```

```python
fig, ax = plt.subplots()

# Add the Seattle temperature data in each month with standard deviation error bars
ax.errorbar(seattle["DATE"], seattle["MLY-TAVG-NORMAL"], yerr=seattle["MLY-TAVG-STDDEV"])

# Set the y-axis label
ax.set_ylabel("Temperature (Fahrenheit)")

plt.show()
```

```python
fig, ax = plt.subplots()

# Add a boxplot for the "Height" column in the DataFrames
ax.boxplot([mens_rowing["Height"]])

# Add x-axis tick labels:
ax.set_xticklabels(["Rowing"])

# Add a y-axis label
ax.set_ylabel("Height (cm)")

plt.show()
```

```python
fig, ax = plt.subplots()

# Add data: "co2" on x-axis, "relative_temp" on y-axis
ax.scatter(climate["co2"], climate["relative_temp"])

# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel("CO2 (ppm)")

# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel("Relative temperature (C)")

plt.show()
```

```python
fig, ax = plt.subplots()

# Add data: "co2", "relative_temp" as x-y, index as color
ax.scatter(climate["co2"], climate["relative_temp"], c=climate.index)

# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel("CO2 (ppm)")

# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel("Relative temperature (C)")

plt.show()
```
